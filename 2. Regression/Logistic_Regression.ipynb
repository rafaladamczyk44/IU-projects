```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

### Dataset

```python
x1 = np.random.normal(size=150)
y1 = (x1 > 0).astype(float)
x1[x1 > 0] *= 4
x1 += 0.3 * np.random.normal(size=150)
x1 = x1.reshape(-1, 1)
```

```python
plt.scatter(x1, y1, color='red')
plt.ylabel('Y')
plt.xlabel('X')
```

```python
lm_log = LogisticRegression().fit(x1, y1)
```

```python
x1_sorted = np.sort(x1, axis=0)
plt.scatter(x1.ravel(), y1, color='black', zorder=20, alpha=0.5)
plt.plot(x1_sorted, lm_log.predict_proba(x1_sorted)[:,1], color='blue' , linewidth = 3)
plt.ylabel('Target')
```
